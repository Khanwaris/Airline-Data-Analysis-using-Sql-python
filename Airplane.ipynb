```python
total_revenue="select a.flight_id,a.flight_no,a.aircraft_code,b.total_seats,coalesce(c.booked_seats,0)as booked_seats,round((COALESCE(c.booked_seats, 0)*1.0/b.total_seats)*100,2) as occupancy from(select flight_id,flight_no,aircraft_code from flights) a left join(select aircraft_code,count(seat_no)as total_seats from seats group by aircraft_code)b on a.aircraft_code=b.aircraft_code left join(select flight_id,count(seat_no)as booked_seats from boarding_passes group by flight_id)c on a.flight_id=c.flight_id" 
occupancy=pd.read_sql(total_revenue,conn)
occupancy
```

```python
low_occupancy="select a.flight_id,a.flight_no,a.aircraft_code,b.total_seats,coalesce(c.booked_seats,0)as booked_seats,round((COALESCE(c.booked_seats, 0)*1.0/b.total_seats)*100,2) as occupancy from(select flight_id,flight_no,aircraft_code from flights) a left join(select aircraft_code,count(seat_no)as total_seats from seats group by aircraft_code)b on a.aircraft_code=b.aircraft_code left join(select flight_id,count(seat_no)as booked_seats from boarding_passes group by flight_id)c on a.flight_id=c.flight_id order by occupancy asc" 
lo=pd.read_sql(low_occupancy,conn)
lo
```

```python
city_highest="SELECT f.departure_airport,f.arrival_airport,SUM(tf.amount) AS total_revenue FROM flights f JOIN ticket_flights tf ON f.flight_id = tf.flight_id GROUP BY f.departure_airport,f.arrival_airport ORDER BY total_revenue DESC LIMIT 10"
city_highest=pd.read_sql(city_highest,conn)
city_highest['route'] = city_highest['departure_airport'] + " → " + city_highest['arrival_airport']
```

```python
city_lowest="SELECT f.departure_airport,f.arrival_airport,SUM(tf.amount) AS total_revenue FROM flights f JOIN ticket_flights tf ON f.flight_id = tf.flight_id GROUP BY f.departure_airport,f.arrival_airport ORDER BY total_revenue asc LIMIT 10"
city_lowest=pd.read_sql(city_lowest,conn)
city_lowest['route'] = city_lowest['departure_airport'] + " → " + city_lowest['arrival_airport']
```

```python
plt.figure()

plt.bar(city_highest['route'].astype(str), city_highest['total_revenue'])

plt.xlabel("Departure")
plt.ylabel("Total revenue")
plt.title("top 10 Routes by Revenue")

plt.xticks(rotation=70)

plt.show()
```

```python
plt.figure()

plt.bar(city_lowest['route'].astype(str), city_lowest['total_revenue'])

plt.xlabel("Departure")
plt.ylabel("Total revenue")
plt.title("lowest 10 Routes by Revenue")

plt.xticks(rotation=70)

plt.show()
```
